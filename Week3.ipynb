```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

print('-----------------------------------')
print('      Mengubah tingkat kecerahan')
print('-----------------------------------')

# Input nilai kecerahan dari pengguna
try:
    brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, not a number')

# Path gambar
image_path = '/content/drive/MyDrive/PCVK_4K/pocong.jpeg'

# Baca gambar
original = cv2.imread(image_path)

# Cek apakah gambar berhasil dibaca
if original is None:
    print("Gambar tidak ditemukan atau tidak bisa dibaca.")
else:
    # Buat array kosong untuk gambar hasil perubahan brightness
    brightness_image = np.zeros(original.shape, original.dtype)

    # Akses per piksel dan ubah kecerahan
    for y in range(original.shape[0]):
        for x in range(original.shape[1]):
            for c in range(original.shape[2]):
                brightness_image[y, x, c] = np.clip(original[y, x, c] + brightness, 0, 255)

    # Cara lebih simple tanpa for loop (alternatif)
    # brightness_image = cv2.convertScaleAbs(original, beta=brightness)

    # Gabungkan gambar asli dan gambar dengan kecerahan baru
    final_frame = cv2.hconcat([original, brightness_image])

    # Tampilkan hasil menggunakan cv2_imshow di Colab
    cv2_imshow(final_frame)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Fungsi untuk mengubah kecerahan dan kontras
def adjust_brightness_contrast(image, brightness=0, contrast=1.0):
  new_image = np.clip(image + brightness, 0, 255).astype(np.uint32)
  new_image = np.clip(((new_image - 127.5) * contrast + 127.5), 0, 255).astype(np.uint8)
  return new_image

# Baca gambar
mandrill_img = cv2.imread('/path/to/mandrill_image.jpg')

# Atur kecerahan dan kontras untuk gambar Mandrill
brightness_mandrill = 50  # Ubah sesuai kebutuhan
contrast_mandrill = 1.5   # Ubah sesuai kebutuhan
mandrill_adjusted = adjust_brightness_contrast(mandrill_img, brightness_mandrill, contrast_mandrill)

# Gabungkan gambar asli dan gambar yang diubah
mandrill_result = cv2.hconcat([mandrill_img, mandrill_adjusted])

# Tampilkan hasilnya
cv2_imshow(mandrill_result)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Fungsi untuk mengubah kecerahan dan kontras
def adjust_brightness_contrast(image, brightness=0, contrast=1.0):
    new_image = np.clip(image + brightness, 0, 255).astype(np.uint8)
    new_image = np.clip(((new_image - 127.5) * contrast + 127.5), 0, 255).astype(np.uint8)
    return new_image

# Baca gambar
flower_woman_img = cv2.imread('/path/to/flower_woman_image.jpg')

# Atur kecerahan dan kontras untuk gambar wanita
brightness_woman = 50  # Sesuaikan nilai kecerahan
contrast_woman = 2.0   # Sesuaikan nilai kontras
woman_adjusted = adjust_brightness_contrast(flower_woman_img, brightness_woman, contrast_woman)

# Gabungkan gambar asli dan gambar yang diubah
woman_result = cv2.hconcat([flower_woman_img, woman_adjusted])

# Tampilkan hasilnya
cv2_imshow(woman_result)
```
